```python
import pandas as pd
import ast
```

```python
data = pd.read_csv('Final_Eval.csv')
data.head()
```

```python
for col in data.columns:
    if 'score' in col:
        data.drop(columns=[col], inplace=True)
```

```python
for col in data.columns:
    if '_eval' in col:
        tmp = []

        for i in range(len(data[col])):


            value = data[col][i]

            if len(value) > 50:
                value = value.replace('\n', '').replace(' ', '')
                value = ast.literal_eval(value)
                tmp.append(value)

            else:
                tmp.append(None)

        data[col] = tmp
        print(col)
```
