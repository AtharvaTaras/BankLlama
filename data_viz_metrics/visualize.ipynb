```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Expanded.csv')
```

```python
new_cols = []

for col in data.columns:
    if 'llama2_shard' in col:
        col = col.replace('llama2_shard', 'llama2_shard\nOUR MODEL')
        
    new_cols.append(col)
    
data.columns = new_cols
```

```python
import matplotlib.pyplot as plt

# Assuming data is already defined
models = data.columns[1:8]
metrics = ['_evalanswer_correctness', '_evalanswer_length', '_evalanswer_formatting', '_evalcontext_understanding']

# Assuming params is defined and contains the parameters for each model

# Get the score of the third model for normalization
third_model_score = None

for param in params:
    if 'model[3]' in param:
        third_model_score = sum(data[param])
        break

if third_model_score is None:
    print("Third model's score not found.")
    exit()

for metric in metrics:

    values = []
    labels = []

    for param in params:
        if metric in param:
            score = sum(data[param])
            # Calculate percentage compared to the third model
            percentage = (score / third_model_score) * 100
            values.append(percentage)
            labels.append(param.replace(metric, ''))

    plt.figure(figsize=(10, 6))
    plt.bar(labels, values, color='skyblue')

    plt.title(f'Bar Plot for {metric.replace("_eval", "").upper()}')
    plt.xlabel('Model Name')
    plt.ylabel('Percentage Compared to Model[3]')
    plt.xticks(rotation=45, ha='right')

    # Adding value labels on top of each bar
    for i in range(len(labels)):
        plt.text(labels[i], values[i], f'{values[i]:.2f}%', ha='center', va='bottom')

    # Display plot
    plt.tight_layout()
    plt.savefig(f'bar_plot_{metric}.jpeg', format='jpeg')
    plt.show()
```
