```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Eval_Cleaned.csv')
data.head()
```

```python
params = ['answer_correctness', 'answer_length', 'answer_formatting', 'context_understanding']
```

```python
data.columns
```

```python
for col in data.columns:
    if '_eval' in col:
        for param in params:
            tmp = []

            for i in range(len(data[col])):
                value = data[col][i]
                value = (eval(value)[param])
                tmp.append(value)

            colname = col + param
            data[colname] = tmp
```
